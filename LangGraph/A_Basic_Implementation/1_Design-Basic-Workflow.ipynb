## **Example 01: Simple WorkFlow**

```python
# Create Function or Component or node:

def function_1(input_1):
    return input_1 + " First Function --> "

def function_2(input_2):
    return input_2 + "to Second Function"
```

```python
# Now Create a Workflow using LangGraph:
from langgraph.graph import Graph

workflow = Graph() # name the workflow

# Add nodes to Workflow:
workflow.add_node("Node_1", function_1)
workflow.add_node("Node_2", function_2)

# Add Edges (data flow between two nodes):
workflow.add_edge("Node_1", "Node_2")

# Set an Entry Point or Start Point:
workflow.set_entry_point("Node_1")

# Set an Exist Point or End Point:
workflow.set_finish_point("Node_2")



# Render or Compile the graph:
app = workflow.compile()
```

```python
app
```

```python
# Invoke:

app.invoke("I am moving from")
```

```python
# Stream:

input = 'I am moving from'
for output in app.stream(input):
    # stream() yields dictionaries with output keyed by node name
    for key, value in output.items():
        print(f"Output from node '{key}':")
        print("---")
        print(value)
    print("\n---\n")
```

## **Example 03: Create Complex WorkFlow:**

* Create Multiple Nodes,
* Nodes contain multiple functions.

```python
def info(dct:dict):
  return dct
```

```python
# Create Nodes:

def function1(dct):
  dct = info(dct=dct)
  return dct

# Convert to Upper
def function2(inputs):
  dct = dd = { k:str(v).upper() for k, v in inputs.items()}
  return dct
```

```python
# Create Workflow:
from langgraph.graph import Graph

workflow = Graph() # name the workflow

# Add Nodes to WorkFlow:
workflow.add_node("Inputs", function1)
workflow.add_node("Upper", function2)

# Add Edges (data flow between nodes):
workflow.add_edge("Inputs", "Upper")

# Set Entry Point:
workflow.set_entry_point("Inputs")

# Set Exist Point:
workflow.set_finish_point("Upper")


# Render or Compile the workflow:
app = workflow.compile()
app
```

```python
app.invoke({"fname":"Dibyendu", 'lname':'Biswas', "age":25})
```

## **Example 04: Create Complex Graph using Multiple Nodes:**

```python
# Create Nodes:

# Node1: Get the User Query as Dictionay Format
def get_user_query(dct:dict):
  return dct


# Node 2: Apply Filter to get specific information (if present):
def filter(dct: dict):
    filtered_data = {}
    keys_to_extract = ["fname", "lname", "age", "country", "address", "state", "pincode", "phone"]
    for key in keys_to_extract:
        if key in dct:
            if key == "fname" or key == "lname":
                if "fullname" not in filtered_data:
                    filtered_data["fullname"] = ""
                filtered_data["fullname"] += " " + str(dct[key])
            else:
                filtered_data[key] = dct[key]

    if "fname" in dct and "lname" in dct:
      filtered_data["fullname"] = filtered_data["fullname"].strip()
    return filtered_data


# Node 3: Convert Into UPPER CASE:
def convert_to_upper(dct:dict):
  new_dct = {}
  for k, v in dct.items():
    if isinstance(v, str):
      new_dct[k] = str(v).upper()
    else:
      new_dct[k] = v
  return new_dct
```

```python
# Create WorkFlow:
from langgraph.graph import Graph

workflow = Graph() # name the workflow

# Add Nodes to Workflow:
workflow.add_node("User Query", get_user_query)
workflow.add_node("Filter", filter)
workflow.add_node("Upper", convert_to_upper)

# Add Edges (data flow between nodes):
workflow.add_edge("User Query", "Filter")
workflow.add_edge("Filter", "Upper")

# Set Entry Point:
workflow.set_entry_point("User Query")

# Set Exist Point:
workflow.set_finish_point("Upper")


# Render or Compile the workflow:
app = workflow.compile()
app
```

```python
app.invoke(
    {
        "fname": "Dibyendu",
        "lname": "Biswas",
        "age": 25,
        "ajaja": 181118818181818181818181,
        "address": "Siliguri",
        "Okay": "buye!!!",
        "city": "Siliguri",
        "state": "West Bengal",
        "country": "India",
        "pincode": 7340001,
        "Hello": "Hello",
        "Instagram": "no no jonnoononono",
        "tata": 'by by'
    }
)
```
